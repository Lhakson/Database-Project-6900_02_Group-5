```python
import pandas as pd
import numpy as np

df = pd.read_csv('All Recorded Traffic.txt',sep=None, engine='python')
df.head(5)
```

```python
df['TIMESTAMP'] = pd.to_datetime(df['DATE']) + pd.to_timedelta(df['TIME'].astype(int), unit='h')
```

```python
df['MAINTENANCE_FLAG'] = np.where(df['LANEMODE'] == 'D', 0, 1)
```

```python
def categorize_fac(name):
    name = str(name).upper()
    if 'TUNNEL' in name: return 'Tunnel'
    if 'BRIDGE' in name or 'CROSSING' in name: return 'Bridge'
    return 'Others'        
```

```python
df['FAC_TYPE'] = df['FAC_B'].apply(categorize_fac)
```

```python
df['VIOLATION_RATE'] = np.where(df['TOTAL'] > 0, (df['VIOLATION'] / df['TOTAL']) * 100, 0)
df.head(5)
```

```python
output_path = 'PA_Traffic__Clean.csv'
df.to_csv(output_path, index=False)
```
