import pandas as pd
import numpy as np

df = pd.read_csv('All Recorded PABT Bus.txt',sep=None, engine='python')
df.head(5)

```python
df.columns = ['Start_Date', 'End_Date', 'Carrier', 'Volume']
```

```python
df['Start_Date'] = pd.to_datetime(df['Start_Date']).dt.date
df['End_Date'] = pd.to_datetime(df['End_Date']).dt.date
```

```python
# Create a mapping for all your identified duplicates
name_mapping = {
    'NJTRANSIT': 'NJ TRANSIT',
    'NJ TRANSIT': 'NJ TRANSIT',
    'C & J BUS LINES': 'C&J BUS LINES',
    'C&J BUS LINES': 'C&J BUS LINES',
    # Academy had two entries in your list; this merges them
    'ACADEMY': 'ACADEMY'
}

# 1. Standardize to Uppercase and Strip whitespace first
df['Carrier'] = df['Carrier'].str.upper().str.strip()

# 2. Apply the mapping
df['Carrier'] = df['Carrier'].replace(name_mapping)

# 3. Re-run the sum to see the true totals
carrier_totals_cleaned = df.groupby('Carrier')['Volume'].sum().sort_values(ascending=False)
print(carrier_totals_cleaned)
```

```python
null_count = df['Volume'].isna().sum()
null_count
```

```python
df.to_csv('PABT_Bus_Cleaned.csv', index=False)
```
